```python
def test(location="New York", profession="Data Engineer", responses=[], start=0, end=1000):
    job_postings = []
    start = start # 1000 is the limit
    while start<end:
        time.sleep(randint(0, 5) * 1)
        print(f"Testing {start}")
        try:
            url = (
                f"https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search/?keywords={profession}"
                + (f"&location={location}" if location else "")
                + f"&start={start}"
            )
            response = requests.get(url, timeout=10)
            response.raise_for_status()
            responses.append(response.text)
            result = Scraper().search_results_parsing(response.text)
            job_postings.extend(result)
            start += 25
        except requests.HTTPError as err:
            print(f"[!!] Something went wrong at {start}! {err}")
            break
    return job_postings    
```

```python
cursor = 0
for result in soup.find_all("li"):
    salary_info = result.find("span", {"class": "job-search-card__salary-info"})
    if salary_info:
        print(re.sub(re.compile(r"\s+"), "", salary_info.text))
    else:
        company_name = result.find(
                "h4", {"class": "base-search-card__subtitle"}
            ).text.strip()
        print(f"Number {cursor} went wrong, company name: {company_name}")
    cursor += 1
```

# Data ingest

```python
from sqlalchemy import create_engine
from sqlalchemy import text
db = "jobs_database"
table_name = "job_search_results"
postgres_url = f'postgresql+psycopg2://root:root@localhost:5432/jobs_database'
engine = create_engine(postgres_url)
```

```python
with engine.connect() as conn:
    result = conn.execute(text(f"SELECT * FROM {table_name} LIMIT 5;")).fetchall()
    for row in result:
        print(row)
```
